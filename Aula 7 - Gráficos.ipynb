```python
data.head(5)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 6)
sns.boxplot(ax=ax, data=data)
sns.despine()
```

```python
sns.boxplot(data=data["diastolic_bp"])
```

```python
# Normalizando
data_norm = data
```

```python
data_norm["id"] = data["id"].div(data["id"].sum(), axis=0)
data_norm["num_preg"] = data["num_preg"].div(data["num_preg"].sum(), axis=0)
data_norm["glucose_conc"] = data["glucose_conc"].div(data["glucose_conc"].sum(), axis=0)
data_norm["diastolic_bp"] = data["diastolic_bp"].div(data["diastolic_bp"].sum(), axis=0)
data_norm["insulin"] = data["insulin"].div(data["insulin"].sum(), axis=0)
data_norm["bmi"] = data["bmi"].div(data["bmi"].sum(), axis=0)
data_norm["diab_pred"] = data["diab_pred"].div(data["diab_pred"].sum(), axis=0)
data_norm["age"] = data["age"].div(data["age"].sum(), axis=0)
data_norm["skin"] = data["skin"].div(data["skin"].sum(), axis=0)
data_norm["diabetes"] = data["diabetes"].div(data["diabetes"].sum(), axis=0)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 6)
sns.boxplot(ax=ax, data=data_norm)
sns.despine()
```

### Mapas de Calor

```python
voos = sns.load_dataset("flights")
voos.head(5)
```

```python
voos_pivoted = voos.pivot("month", "year", "passengers")
voos_pivoted.head(5)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18, 8)
sns.heatmap(voos_pivoted)
```

```python
plt.subplots(figsize=(18, 8))
sns.heatmap(voos_pivoted, annot=True, fmt="d")
```

```python
plt.subplots(figsize=(18, 8))
sns.heatmap(voos_pivoted, annot=True, fmt="d", center=voos_pivoted.loc["January", 1955])
```

```python
sns.clustermap(voos_pivoted, figsize=(10,10))
```
